```python
import scipy
def linear_reg (X,Y):
    X_log = np.log(X)
    Y_log = np.log(Y)
    b, log_a, r_value, p_value, std_err = scipy.stats.linregress(X_log, Y_log)
    a = np.exp(log_a)
    x_min = np.min(X)
    y_min = a*(x_min**b)
    x_max = np.max(X)
    y_max = a*(x_max**b)
    return a, b, r_value**2, [x_min, x_max], [y_min, y_max]
```

```python
def k_b(X, Y):
    k = (Y[1] - Y[0]) / (-X[0] + X[1])
    b = -k * X[0] + Y[0]
    return k, b

def x_y_peresech(K, B):
    x = (B[1] - B[0])/(K[0] - K[1])
    y = K[0]*x + B[0]
    return x,y


def OFP(plot_data):
    Kov = plot_data['x'].min()
    Kv_1 = plot_data[plot_data['OFP_w']==0]['x'].max()
    Kv_2 = plot_data[plot_data['OFP_n']==0]['x'].min()
    
    index_false = list(plot_data['OFP_n'] > plot_data['OFP_w']).index(False)
    index_true = index_false-1
    data_search = plot_data.iloc[[index_false, index_true]]
    
    x = np.array(data_search['x'])
    y1 = np.array(data_search['OFP_w'])
    y2 = np.array(data_search['OFP_n'])
    
    K = [k_b(x, y1)[0], k_b(x, y2)[0]]
    B = [k_b(x, y1)[1], k_b(x, y2)[1]]

    Kv_kr = x_y_peresech(K,B)[0]
    return Kov, Kv_1,Kv_kr, Kv_2 
```

```python
import scipy
def linear_reg (X,Y):
    a, b, r_value, p_value, std_err = scipy.stats.linregress(X, Y)
    x_min = np.min(X)
    y_min = a*x_min+b
    x_max = np.max(X)
    y_max = a*x_max+b
    return a, b, r_value**2, [x_min, x_max], [y_min, y_max]
```

```python
import scipy
def pow_reg (X,Y):
    X_log = np.log(X)
    Y_log = np.log(Y)
    b, log_a, r_value, p_value, std_err = scipy.stats.linregress(X_log, Y_log)
    a = np.exp(log_a)
    x_min = np.min(X)
    y_min = a*(x_min**b)
    x_max = np.max(X)
    y_max = a*(x_max**b)
    return a, b, r_value**2, [x_min, x_max], [y_min, y_max]
    
def exp_reg (X,Y):
    X_log = X
    Y_log = np.log(Y)
    b, exp_a, r_value, p_value, std_err = scipy.stats.linregress(X_log, Y_log)
    a = np.exp(exp_a)
    x_min = np.min(X)
    y_min = a*np.exp(x_min*b)
    x_max = np.max(X)
    y_max = a*np.exp(x_max*b)
    return a, b, r_value**2, [x_min, x_max], [y_min, y_max]

def log_reg (X,Y):
    X_log = np.log(X)
    Y_log = Y
    b, a, r_value, p_value, std_err = scipy.stats.linregress(X_log, Y_log)
    x_min = np.min(X)
    y_min = a*np.exp(x_min*b)
    x_max = np.max(X)
    y_max = a*np.exp(x_max*b)
    return a, b, r_value**2, [x_min, x_max], [y_min, y_max]

def linreg(X,Y):
    a, b, r_value, p_value, std_err = scipy.stats.linregress(X, Y)
    x_min = np.min(X)
    y_min = (x_min)*a+b
    x_max = np.max(X)
    y_max = (x_max)*a+b
    return b, a, r_value**2, [x_min, x_max], [y_min, y_max]
    
```

```python
def max_pow_reg(X_not_node, Y_not_node):
    def check_and_remove_top_row(datafra):
        # Проверяем корреляцию
        a, b, r_value, [x_min, x_max], [y_min, y_max] = pow_reg(datafra.X, datafra.Y)
        # Выводим матрицу корреляции
    
        # Удаляем верхнюю строку
        datafra.drop(datafra.index[0], inplace=True)
        rez = [a, b, r_value, [x_min, x_max], [y_min, y_max]]
        # Возвращаем обновленный датафрейм
        return datafra, r_value, rez
    
    # Применяем функцию
    XY_not_node = pd.DataFrame({'X':X_not_node,'Y':Y_not_node})
    XY_not_node = XY_not_node.sort_values(by='Y')
    rez_list = []
    max_len = len(XY_not_node)/2
    while len(XY_not_node) > max_len:
        XY_not_node, r_value, rez= check_and_remove_top_row(XY_not_node)
        rez_list.append([r_value, rez])
    rez_list = pd.DataFrame(rez_list)
    rez_list = rez_list.nlargest(1, columns = 0)[1]
    return rez_list
```

```python
def k_b(X, Y):
    k = (Y[1] - Y[0]) / (-X[0] + X[1])
    b = -k * X[0] + Y[0]
    return k, b

def x_y_peresech(K, B):
    x = (B[1] - B[0]) / (K[0] - K[1])
    y = K[0] * x + B[0]
    return x, y
    
def OFP(plot_data):
    Kov = plot_data['x'].min()
    Kv_1 = plot_data[plot_data['OFP_w'] == 0]['x'].max()
    Kv_2 = plot_data[plot_data['OFP_n'] == 0]['x'].min()

    index_false = list(plot_data['OFP_n'] > plot_data['OFP_w']).index(False)
    index_true = index_false - 1
    data_search = plot_data.iloc[[index_false, index_true]]

    x = np.array(data_search['x'])
    y1 = np.array(data_search['OFP_w'])
    y2 = np.array(data_search['OFP_n'])

    K = [k_b(x, y1)[0], k_b(x, y2)[0]]
    B = [k_b(x, y1)[1], k_b(x, y2)[1]]

    Kv_kr = x_y_peresech(K, B)[0]
    return Kov, Kv_1, Kv_kr, Kv_2
def pow_reg(X, Y):
    X_log = np.log(X)
    Y_log = np.log(Y)
    b, log_a, r_value, p_value, std_err = scipy.stats.linregress(X_log, Y_log)
    a = np.exp(log_a)
    x_min = np.min(X)
    y_min = a * (x_min ** b)
    x_max = np.max(X)
    y_max = a * (x_max ** b)
    return a, b, r_value ** 2, [x_min, x_max], [y_min, y_max]
```
